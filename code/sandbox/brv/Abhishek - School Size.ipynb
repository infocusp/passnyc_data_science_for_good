```python
for ix, row in df_school_detail.iterrows():
    csd = str(row['CSD'])
    if len(csd) == 1:
        csd = "0"+csd
        df_school_detail.loc[ix,'CSD'] = csd
```

```python
df_school_detail["DBN_manual"] = (df_school_detail["CSD"].astype(str)) + (df_school_detail["SCHOOL CODE"].astype(str)) 
```
